```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data as a dictionary
data={
    'bedrooms':hii['bedrooms'],
    'bed_type':hii['bed_type']
}
# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Group by 'bedrooms' and 'bed_type', then count the occurrences
grouped_data = df.groupby(['bedrooms', 'bed_type']).size().unstack(fill_value=0)

# Plot the bar chart
ax = grouped_data.plot(kind='bar', figsize=(10,5))

# Set x-axis limit

# Add labels and title
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count of Bed Types')
plt.title('Bedrooms vs Bed Type')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]
# Create a pie chart with a hole in the center (donut chart)
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, wedgeprops={'width': 1})
# Add a circle at the center to create the hole
centre_circle = plt.Circle((0,0),0.4,fc='white')
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  
# Add title
plt.title('Donut Chart Example')
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming `hii` is your DataFrame and it contains the 'bed_type' column
# Get unique values and their counts
bed_type_counts = hii['bathrooms'].value_counts()

# Create a new DataFrame with 'bed_type' and 'count' columns
bed_type_df = bed_type_counts.reset_index()
bed_type_df.columns = ['bathrooms', 'count']

# Print the new DataFrame
print(bed_type_df)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(bed_type_df['bathrooms'], bed_type_df['count'])
plt.xlabel('Bed Type')
plt.ylabel('Count')
plt.title('Count of Each Bed Type')
plt.xticks(rotation=45)
plt.show()
```
